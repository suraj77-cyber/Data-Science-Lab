4) Data Viz:
Histograms show how numerical values are distributed, while frequency plots show how often values occur. Visualization makes patterns and trends easier to understand compared to raw tables.

```python
num_cols = df.select_dtypes(include=['int64','float64']).columns
num_cols
```

```python
import matplotlib.pyplot as plt

for col in num_cols:
    plt.figure()
    plt.hist(df[col].dropna())
    plt.title(f"Histogram of {col}")
    plt.show()
```

```python
plt.figure()
df[col].value_counts().head(10).plot(kind='bar')
plt.title(f"Top 10 Frequency Plot of {col}")
plt.xticks(rotation=45)
plt.show()
```
