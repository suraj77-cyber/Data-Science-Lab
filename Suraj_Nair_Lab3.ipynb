```python
## Matplotlib.pyplot

import matplotlib.pyplot as plt

x = np.arange(0,3*np.pi,0.1)
y_sin= np.sin(x)
y_cos= np.cos(x)

plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('Variable X')
plt.ylabel('Trignomtric Function')
plt.title('Sine and Cosine')
plt.show()
```

```python
## Subplot
plt.subplot(2,1,1)
plt.plot(x,y_sin,color = "green")
plt.plot(x,y_sin)
plt.xlabel('Variable X')
plt.ylabel('Sin Function')
plt.title('Trignometric_Sin Function')

plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.xlabel('Variable X')
plt.ylabel('Cos Function')
plt.title('Trignometric_Cos Function')
plt.show()

```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 10, 0.1)

y_linear = 2*x + 1          # Linear
y_power = x**2              # Power
y_exponential = 2**x        # Exponential
y_logarithmic = np.log(x)   # Logarithmic (natural log)

plt.figure(figsize=(10, 8))

## Linear Function
plt.subplot(2, 2, 1)
plt.plot(x, y_linear, color='blue')
plt.title("Linear Function")
plt.grid(True)

# Power Function
plt.subplot(2, 2, 2)
plt.plot(x, y_power, color='green')
plt.title("Power Function")
plt.grid(True)

#Exponential Function
plt.subplot(2, 2, 3)
plt.plot(x, y_exponential, color='red')
plt.title("Exponential Function")
plt.grid(True)

#Logarithmic Function
plt.subplot(2, 2, 4)
plt.plot(x, y_logarithmic, color='purple')
plt.title("Logarithmic Function")
plt.grid(True)

plt.show()
```
